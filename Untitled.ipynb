```python
director_data = {
    'Director': ['Alonso Ramirez Ramos', 'John Cherry', 'Karen Disher', 'Hamish Hamilton', 'Nan'],
    'Popularity': [90, 88, 85, 82, 80]  # Les valeurs de popularitÃ© sont fictives
}

# Exemple de donnÃ©es fictives pour les acteurs
actor_data = {
    'Actor': ['Chris Diamantopoulos', '', 'Jim Varney', 'Noelle Parker', 'Douglas Seal', 'Mayur More', 'Mark Ruffalo', 'Jesse Eisenber', 'Tim Allen', 'Vicky Kaushalp'],
    'Rating': [95, 92, 90, 88, 87, 85, 84, 82, 80, 78]  # Les valeurs de notation sont fictives
}

# CrÃ©er des DataFrames Ã  partir des donnÃ©es
top_directors = pd.DataFrame(director_data)
top_actors = pd.DataFrame(actor_data)

# Tracer des graphiques Ã  barres pour les rÃ©alisateurs et les acteurs
plt.figure(figsize=(12, 6))

# Top 5 des rÃ©alisateurs
plt.subplot(1, 2, 1)
sns.barplot(x='Director', y='Popularity', data=top_directors, palette='magma')
plt.title('Top 5 des rÃ©alisateurs')
plt.xlabel('RÃ©alisateur')
plt.ylabel('PopularitÃ©')
plt.xticks(rotation=45)

# Top 10 des acteurs
plt.subplot(1, 2, 2)
sns.barplot(x='Actor', y='Rating', data=top_actors, palette='magma')
plt.title('Top 10 des acteurs')
plt.xlabel('Acteur')
plt.ylabel('Notation')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python

tv_series_data = {
    'Title': ['The Beatles:get back', 'Hawkeye	', 'port protection alaska', 'secrets of the zoo:tampa', 'Dr: Oakley ,yukon vet'],
    'Seasons': [1, 1, 2, 2, 2]
}

# CrÃ©er un DataFrame Ã  partir des donnÃ©es
tv_series = pd.DataFrame(tv_series_data)

# Trier les sÃ©ries tÃ©lÃ©visÃ©es par le nombre de saisons (du plus grand au plus petit)
tv_series_sorted = tv_series.sort_values(by='Seasons', ascending=False).head(5)

# Tracer un graphique Ã  barres pour les cinq meilleures sÃ©ries tÃ©lÃ©visÃ©es avec le plus grand nombre de saisons
plt.figure(figsize=(10, 6))
sns.barplot(x='Title', y='Seasons', data=tv_series_sorted, palette='pastel')
plt.title('Top 5 des sÃ©ries tÃ©lÃ©visÃ©es avec le plus grand nombre de saisons')
plt.xlabel('SÃ©rie tÃ©lÃ©visÃ©e')
plt.ylabel('Nombre de saisons')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Vos textes
texts = [
    "Join Mickel and the gang as they duck the halls...",
    "Santa Claus passes his magic bag to a new St. Nic...",
    "Sid the Sloth is on Santa's naughty list...",
    "This is real life , not just fantasy!...",
    "A three-part documentary from Peter Jacques capturing a moment in music history with the Beatles...",
    "An inside look at the legendry life of adventurer Jacques-yves cousteau ...",
    "Clint Barton/Hawkeye must team up with skilled archer Kate Bishop to unravel a criminal conspiracy...",
    "Resident of port Protection must cobat volatile conditions to survive and thrive in Alaska...",
    "A day in the life at ZooTampa is anything but ordinary. It's extraordunary..." ,
    "Celebrate the holiday season with all favorite Muppets..." ,
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)

# Calcul des similaritÃ©s cosinus
cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Affichage des paires de textes similaires
for i in range(len(texts)):
    for j in range(i+1, len(texts)):
        print(f"SimilaritÃ© entre texte {i+1} et texte {j+1}: {cosine_similarities[i][j]}")
```
